```python
import pandas as pd
```

```python
df = pd.read_csv("contract_recommendation.csv")
```

```python
df.rename(columns={"Unnamed: 0": "player_id"}, inplace=True)
```

```python
label = pd.read_csv("./knowledge/contract_details.csv")
```

```python
label.reset_index(drop=False, inplace=True)
```

```python
label.rename(columns={"index": "player_id"}, inplace=True)
```

```python
df = pd.merge(df, label, on="player_id", suffixes=("_pred", "_true"), how="inner")
```

```python
df['diff']  = abs(df['value_pred']/df['yrs_pred'] - df['average_salary'])
```

```python
df.groupby('tier')['diff'].mean().sort_values(ascending=False)
```
